### weekly sales distribution

```python
df= pd.read_csv("C:\\Users\\HIMANSHI\\OneDrive\\Documents\\walmart_sales.csv")
plt.figure(figsize=(10,5))
plt.hist(df['Weekly_Sales'], bins=50, color='steelblue')
plt.title('Weekly Sales Distribution')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.show()
```

### temperature vs weekly_sales and size vs weekly_sales

### CPI vsweekliy_sales & unemployment VS weekly_sales

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

df['CPI_Group'] = pd.cut(df['CPI'], bins=5)
df['Unemp_Group'] = pd.cut(df['Unemployment'], bins=5)

df.groupby('CPI_Group')['Weekly_Sales'].mean().plot(kind='bar', color='teal', ax=axes[0])
axes[0].set_title('CPI vs Weekly Sales')
axes[0].set_xlabel('CPI Range')
axes[0].set_ylabel('Avg Weekly Sales')
axes[0].tick_params(axis='x', rotation=45)

df.groupby('Unemp_Group')['Weekly_Sales'].mean().plot(kind='bar', color='orange', ax=axes[1])
axes[1].set_title('Unemployment vs Weekly Sales')
axes[1].set_xlabel('Unemployment Range')
axes[1].set_ylabel('Avg Weekly Sales')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Q1: Top 5 Stores by Total Sales"
query = '''
 SELECT Store, SUM(Weekly_Sales) AS Total_Sales
 FROM walmart_sales
 GROUP BY Store
 ORDER BY Total_Sales DESC
 LIMIT 5;
 '''

# Execute and display as a clean table
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
 # Q2: Holiday vs Non-Holiday Average Sales
query='''
 SELECT IsHoliday, AVG(Weekly_Sales) AS Average_Sales
 FROM walmart_sales
 GROUP BY IsHoliday
 ORDER BY Average_Sales DESC;
 '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
#-- Q3: Top 5 Departments by Total Sale
query='''
SELECT Dept, SUM(Weekly_Sales) AS Total_Sales
FROM walmart_sales
GROUP BY Dept
ORDER BY Total_Sales DESC
LIMIT 5;
'''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
#-- Q4: Best Store Type A, B, C
Query='''

 SELECT Type, AVG(Weekly_Sales) AS Average_Sales
 FROM walmart_sales
 GROUP BY Type
 ORDER BY Average_Sales DESC;
 '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
 #Q5: Total Stores in Database
Query='''

 SELECT COUNT(DISTINCT Store) AS Total_Stores
 FROM walmart_sales;
 '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
 #-- Q6: Most Profitable Year
Query='''

 SELECT Year, SUM(Weekly_Sales) AS Total_Sales
 FROM walmart_sales
 GROUP BY Year
 ORDER BY Total_Sales DESC;
    '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q7: How Many Times Did Weekly Sales Go Above 50,000?
Query='''
 SELECT COUNT(*) AS Times_Above_50000
 FROM walmart_sales
 WHERE Weekly_Sales > 50000;
 '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q8: Most Consistent Store
Query='''
SELECT Store,
MAX(Weekly_Sales) AS max_sales,
MIN(Weekly_Sales) AS min_sales,
MAX(Weekly_Sales) - MIN(Weekly_Sales) AS sales_range
    FROM walmart_sales
    GROUP BY Store
    ORDER BY sales_range ASC
    LIMIT 1;
    '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q9: Top 3 Departments on Holidays
Query ='''
 SELECT Dept, SUM(Weekly_Sales) AS total_sales
 FROM walmart_sales
 WHERE IsHoliday = 1
 GROUP BY Dept
 ORDER BY total_sales DESC
 LIMIT 3;
 '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q10: Average Weekly Sales by Store
Query='''

 SELECT Store, AVG(Weekly_Sales) AS AVERAGE_SALES
 FROM walmart_sales
 GROUP BY Store
 ORDER BY AVERAGE_SALES DESC limit 5;
'''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q11: Temperature Effect on Sales
Query='''

 SELECT CASE
         WHEN Temperature >= 80 THEN 'Hot'
         WHEN Temperature BETWEEN 50 AND 79 THEN 'Moderate'
         ELSE 'Cold'
         END AS temp_category,
         AVG(Weekly_Sales) AS avg_sales
     FROM walmart_sales
     GROUP BY temp_category
     ORDER BY avg_sales DESC;
     '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
#-- Q12: Unemployment Effect on Sales
Query='''

 SELECT CASE
        WHEN Unemployment >= 8 THEN 'High'
        WHEN Unemployment BETWEEN 5 AND 7.99 THEN 'Medium'
        ELSE 'Low'
           END AS unemployment_level,
           AVG(Weekly_Sales) AS avg_sales
     FROM walmart_sales
     GROUP BY unemployment_level
     ORDER BY avg_sales DESC;
     '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
 #-- Q13: Fuel Price Effect on Sales
Query ='''

 SELECT CASE
                WHEN Fuel_Price >= 4 THEN 'High'
                ELSE 'Low'
            END AS fuel_price_level,
            AVG(Weekly_Sales) AS avg_sales
     FROM walmart_sales
     GROUP BY fuel_price_level
     ORDER BY avg_sales DESC;
     '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q14: CPI Impact on Sales
Query='''

 SELECT CASE
                WHEN CPI >= 200 THEN 'High CPI'
                WHEN CPI BETWEEN 150 AND 199.99 THEN 'Medium CPI'
                ELSE 'Low CPI'
            END AS cpi_level,
            AVG(Weekly_Sales) AS avg_sales
     FROM walmart_sales
     GROUP BY cpi_level
     ORDER BY avg_sales DESC;
     '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

```python
# -- Q15: Top Department in Store Type B
Query ='''

SELECT Dept, SUM(Weekly_Sales) AS total_sales
 FROM walmart_sales
 WHERE Type = 'B'
 GROUP BY Dept
 ORDER BY total_sales DESC
 LIMIT 1;
 '''
top_dept_df = pd.read_sql(query, engine)
top_dept_df
```

## 📌 Project Overview
### This project analyzes Walmart's weekly sales data to understand
### what factors influence sales performance.

### **Tools Used:** Python, Pandas, Matplotlib, MySQL
### **Goal:** Identify key drivers of weekly sales across stores.

## 🔍 Key Findings

### 1. Best Performing Store
###### - Store 6 had the highest total sales (~$174M)
###### - Store 4 had the highest average weekly sales (~$19,857)

### 2. Store Type Analysis
###### - Type A > Type B > Type C in average weekly sales
###### - Type A stores significantly outperform others

### 3. Holiday Effect (Surprising! 😮)
###### - Non-holiday weeks had HIGHER sales ($13,072 avg)
###### - Holiday weeks slightly lower ($12,850 avg)
###### - Suggests everyday shopping drives more revenue than holiday promotions

### 4. Monthly Trend
###### - December had the highest sales (festive season) 📈
###### - January had the lowest sales

### 5. Year-wise Performance
###### - 2020 was the most profitable year ($1.72B total)
###### - 2021 showed a decline in sales
###### - 2019 was in between

### 6. Economic Factors
###### - Low unemployment = higher sales ✅
###### - Low CPI (inflation) = higher sales ✅
###### - Fuel price had minimal impact on sales

### 7. Temperature Effect
###### - Moderate temperature (50–79°F) had the highest sales
###### - Extreme hot or cold weather reduced shopping

## ✅ Conclusions & Recommendations

###### 1. **Invest in Type A stores** — they consistently generate the most revenue
###### 2. **Plan inventory for December** — sales peak during festive season
###### 3. **Monitor economic indicators** — unemployment and CPI directly impact sales
###### 4. **Improve holiday promotions** — holiday weeks are underperforming vs non-holiday
###### 5. **Focus on moderate climate regions** — better foot traffic and sales

```python
# Final Summary Dashboard - 4 Key Insights
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Walmart Sales Analysis - Key Insights Summary', 
             fontsize=16, fontweight='bold')

# 1. Store Type Comparison
df.groupby('Type')['Weekly_Sales'].mean().plot(
    kind='bar', ax=axes[0,0], color=['green','orange','red'])
axes[0,0].set_title('Average Sales by Store Type')
axes[0,0].set_ylabel('Avg Weekly Sales')
axes[0,0].set_xlabel('Store Type')

# 2. Monthly Sales Trend
df.groupby('Month')['Weekly_Sales'].mean().plot(
    kind='line', ax=axes[0,1], color='purple', marker='o')
axes[0,1].set_title('Monthly Sales Trend')
axes[0,1].set_xlabel('Month')
axes[0,1].set_ylabel('Avg Weekly Sales')
axes[0,1].grid(True)

# 3. Holiday vs Non-Holiday
df.groupby('IsHoliday')['Weekly_Sales'].mean().plot(
    kind='bar', ax=axes[1,0], color=['grey','pink'])
axes[1,0].set_title('Holiday vs Non-Holiday Sales')
axes[1,0].set_xticklabels(['Non-Holiday', 'Holiday'], rotation=0)
axes[1,0].set_ylabel('Avg Weekly Sales')

# 4. Year-wise Total Sales
df.groupby('Year')['Weekly_Sales'].sum().plot(
    kind='bar', ax=axes[1,1], color='steelblue')
axes[1,1].set_title('Total Sales by Year')
axes[1,1].set_ylabel('Total Sales')
axes[1,1].set_xlabel('Year')

plt.tight_layout()
plt.savefig('walmart_insights_summary.png')  # Saves the image!
plt.show()
print("Summary dashboard saved!")
```
